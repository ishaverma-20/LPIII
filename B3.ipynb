```python
import pandas as pd
df = pd.read_csv("Churn_Modelling.csv")
df
```

```python
df.isnull().sum()
```

```python
df.drop(['RowNumber', 'CustomerId', 'Surname'], axis = 1, inplace = True)
```

```python
df
```

```python
y = df['Exited'].values
x = df.loc[:, df.columns != 'Exited'].values
```

```python
#Encoding and Normalizing Columns Gender and Geography
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

encoder = LabelEncoder()
x[:, 1] = encoder.fit(x[:, 1])
x[:, 2] = encoder.fit(x[:, 2])

pipeline = Pipeline(
    [
        ('Categorizer', ColumnTransformer(
            [
                ("Gender Label Encoder", OneHotEncoder(categories="auto", drop="first"), [2]),
                ("Geography Label Encoder", OneHotEncoder(categories="auto", drop="first"), [1])
            ],
            remainder="passthrough", n_jobs = 1)),
        ('Normalizer', StandardScaler())
    ]
)
```

```python
x = pipeline.fit_transform(x)
x
```

```python
#Splitting data
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.33)
```

```python
#Neural Network
#Creation
from keras.models import Sequential
from keras.layers import Dense, Dropout

classifier = Sequential() #Initialization
```

```python
classifier.add(Dense(6, activation = 'relu', input_shape = (X_train.shape[1], )))
classifier.add(Dropout(rate = 0.1))
```

```python
classifier.add(Dense(6, activation = "relu"))
classifier.add(Dropout(rate = 0.1))
```

```python
classifier.add(Dense(1, activation = "sigmoid"))
```

```python
classifier.summary()
```

```python
#Optimize the weights
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics = ['accuracy'])
```

```python
#Fitting the NN
history = classifier.fit(X_train, Y_train, batch_size = 32, epochs = 50, validation_split = 0.1, verbose = 2)
```

```python
prediction = classifier.predict(X_test)
```

```python
prediction = (prediction>0.5).astype(int)
```

```python
from sklearn import metrics
cm = metrics.confusion_matrix(Y_test, prediction)
print(cm)
```

```python
print(metrics.accuracy_score(Y_test, prediction))
```

```python

print(((cm[0][0] + cm[1][1])* 100) / len(Y_test), '% of data was classified correctly')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar = False,
            xticklabels=["Positive", "Negative"],
            yticklabels=["Positive", "Negative"])

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
