```python
#Implement Gradient Descent Algorithm to find the local minima of a function. For example, find the local minima of the function y=(x+3)Â² starting from the point x=2.
```

```python
import numpy as np
import sympy as sym
from matplotlib import pyplot as plt

# Function to convert sympy expression into a lambda function for evaluation
def lambdify_expression(expr):
    x = sym.symbols('x')
    return sym.lambdify(x, expr, 'numpy')

# Gradient descent function
def gradient(expr, alpha, start, max_iter):
    x = sym.symbols('x')
    grad = sym.Derivative(expr, x).doit()  # Compute symbolic derivative
    objective_fn = lambdify_expression(expr)  # Convert expression to function
    derivative_fn = lambdify_expression(grad)  # Convert derivative to function

    x_list = []
    x_val = start
    x_list.append(x_val)

    for i in range(max_iter):
        gradi = derivative_fn(x_val)  # Compute gradient at current x
        x_val = x_val - (alpha * gradi)  # Update x
        x_list.append(x_val)

    return x_list, objective_fn

# Define the expression (you can change this to any expression)
x = sym.symbols('x')
expr = (sym.exp(x) + x**2)  # Example expression

# Gradient descent parameters
alpha = 0.1
start = 2
max_iter = 30

# Perform gradient descent
x_vals, objective_fn = gradient(expr, alpha, start, max_iter)

# Plotting
x_cor = np.linspace(-3, 3, 100)
plt.plot(x_cor, objective_fn(x_cor), label="Objective function")
plt.plot(start, objective_fn(start), 'ro', label="Starting point")

x_arr = np.array(x_vals)
plt.plot(x_arr, objective_fn(x_arr), '.-', color='red', label="Gradient descent path")
plt.xlabel('x')
plt.ylabel('Objective function value')
plt.legend()
plt.show()

# Final value after gradient descent
print("Final x value:", x_vals[-1])
print("Final objective function value:", objective_fn(x_vals[-1]))
```
